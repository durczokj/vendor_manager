```python
class CostItemAssignment():
    def __init__(
            self,
            undertaking,
            percentage):
        self.undertaking = undertaking
        self.percentage = percentage

class CostItem():
    def __init__(
            self,
            engagement,
            day,
            cost,
            assignments):
        
        self.engagement = engagement
        self.day = day
        self.cost = cost
        self.assignments = assignments

        self.validate()
    
    def validate(self):
        assignment_sum = sum([a.percentage for a in self.assignments])
        assert assignment_sum == 1, "The sum of assignments must be 1."

def list_dates(start_date, end_date):
    d = start_date
    date_list = []
    while d <= end_date:
        date_list.append(d)
        d = d + timedelta(days=1)
    
    return date_list
```
